```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, pipeline, AutoModelForSeq2SeqLM
import gradio as gr
```

```python
model=AutoModelForSequenceClassification.from_pretrained('/content/drive/MyDrive/NN_LLM_Project_Files')
tokenizer=AutoTokenizer.from_pretrained('/content/drive/MyDrive/NN_LLM_Project_Files')
translator=pipeline("translation", model=model, tokenizer=tokenizer)
```

```python
source_text = "Hello, how are you?"
translated_text = translator(source_text, target_language="fr")
print("Translated Text:", translated_text[0]["translation_text"])
```

```python
def translate_text(input_text):
    inputs = tokenizer.encode("translate English to French: " + input_text, return_tensors="pt", max_length=128, truncation=True)
    translated_ids = model.generate(inputs)
    translated_text = tokenizer.decode(translated_ids[0], skip_special_tokens=True)
    return translated_text

iface = gr.Interface(
    fn=translate_text,
    inputs=gr.inputs.Textbox(placeholder="Enter English text..."),
    outputs=gr.outputs.Textbox(label="French translation"),
    live=True,
)

if __name__ == "__main__":
    iface.launch()
```
